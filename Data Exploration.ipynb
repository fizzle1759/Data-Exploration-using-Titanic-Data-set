```python
median_age = df.Age.median()
df["Age"].fillna(median_age,inplace=True)
```

```python
df["Cabin_address"]= df.Cabin.str[0]
```

```python
df["Cabin_address"].mode([0])
```

```python
most_frequent = df["Cabin_address"].mode()[0]

df["Cabin_address"].fillna(most_frequent,inplace=True)
```

```python
most_frequent = df["Embarked"].mode()[0]

df["Embarked"].fillna(most_frequent,inplace=True)
```

```python
split =lambda x : x.split()[1].strip(".").strip(",")
```

```python
d = df.Name.apply(split)

d
```

```python
f = []
for i in d:
    if i not in ["Mr", "Mrs", "Miss", "Master"," Dr", "Rev"]:
        i = "other"
    f.append(i)
```

```python
d.index = f
f
```

```python
df["Title"] = f

df
```

```python
df["family_no"] = df["SibSp"] + df["Parch"]

df
```

```python
df.drop(columns=["Name","SibSp","Parch","Ticket","Cabin"], inplace=True)
```

# Univariate Analysis
exploring variable/column one by one

```python
gender_count= df.Survived.value_counts()
```

```python
gender_count= df.Survived.value_counts()
gender_count.index = ["dead","Survived"] # change the 0 and 1 to dead and survived
```

```python
# pandas plot
gender_count.plot(kind="bar")
plt.title("Count of Dead and Survived",color="red")
plt.ylabel("Count")
plt.xlabel("Survived Rate")
plt.xticks(rotation=30)
plt.show()
```

```python
pclass_count = df.Pclass.value_counts()
pclass_count.index = ["thirdclass", "firstclass","secondclass"]
```

```python
pclass_count.plot(kind="bar")
plt.title("Count of Pclass",color="red")
plt.ylabel("Count")
plt.xlabel("pclass")
plt.xticks(rotation=30)
plt.show()
```

```python
df["Age"].plot(kind="hist")
plt.title("Age Distribution",color="red")
plt.ylabel("Count")
plt.xlabel("Age")
plt.xticks(rotation=30)
plt.show()
```
